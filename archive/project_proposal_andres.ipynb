```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('/Users/andresriera/Documents/Clases UCSD/Junior/Winter 2025/DSC 106/Proqject 4/cases_clean_project 4.csv')

# Drop NaN values in 'optype'
df = df.dropna(subset=['optype'])

# Count occurrences of each surgery type
surgery_counts = df['optype'].value_counts()

# Calculate proportions
surgery_proportions = surgery_counts / surgery_counts.sum()

# Plot pie chart
plt.figure(figsize=(10, 6))
surgery_proportions.plot(kind='pie', autopct='%1.1f%%', startangle=140, cmap='tab10')

# Customize plot
plt.title("Proportion of General Surgeries by Type")
plt.ylabel('')  # Hide y-label for better visualization
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/Users/andresriera/Documents/Clases UCSD/Junior/Winter 2025/DSC 106/Proqject 4/cases_clean_project 4.csv')

# Drop NaN values in 'optype' and 'dx'
df = df.dropna(subset=['optype', 'dx'])

# Define keywords associated with cancer diagnoses
cancer_keywords = ["cancer", "carcinoma", "malignant", "tumor", "neoplasm", "sarcoma", "lymphoma", "leukemia"]

# Function to check if diagnosis contains cancer-related terms
def is_cancer_diagnosis(dx):
    dx_lower = str(dx).lower()  # Convert to lowercase for comparison
    return any(keyword in dx_lower for keyword in cancer_keywords)

# Apply function to determine cancer diagnoses
df['cancer_dx'] = df['dx'].apply(is_cancer_diagnosis)

# Group by 'optype' and calculate proportions of cancer vs non-cancer diagnoses
optype_groups = df.groupby('optype')['cancer_dx'].value_counts(normalize=True).unstack().fillna(0)

# Plot pie charts for each 'optype'
for optype, proportions in optype_groups.iterrows():
    plt.figure(figsize=(6, 6))
    proportions.plot(kind='pie', labels=['Non-Cancer', 'Cancer'], autopct='%1.1f%%', startangle=140, colors=['lightblue', 'red'])
    plt.title(f"Proportion of Cancer Diagnoses for {optype}")
    plt.ylabel('')  # Hide y-label for better visualization
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# Load the dataset
df = pd.read_csv('/Users/andresriera/Documents/Clases UCSD/Junior/Winter 2025/DSC 106/Proqject 4/cases_clean_project 4.csv')

# Drop NaN values in 'optype' and 'dx'
df = df.dropna(subset=['optype', 'dx'])

# Define keywords associated with cancer diagnoses
cancer_keywords = ["cancer", "carcinoma", "malignant", "tumor", "neoplasm", "sarcoma", "lymphoma", "leukemia"]

# Function to check if diagnosis contains cancer-related terms
def is_cancer_diagnosis(dx):
    dx_lower = str(dx).lower()
    return any(keyword in dx_lower for keyword in cancer_keywords)

# Apply function to determine cancer diagnoses
df['cancer_dx'] = df['dx'].apply(is_cancer_diagnosis)

# Group by 'optype' and calculate proportions of cancer vs non-cancer diagnoses
optype_groups = df.groupby('optype')['cancer_dx'].value_counts(normalize=True).unstack().fillna(0)

# Determine grid size (rows and columns)
num_charts = len(optype_groups)
cols = 3  # Number of columns per row
rows = math.ceil(num_charts / cols)  # Calculate needed rows

# Create subplots
fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))  # Adjust figure size

# Flatten axes array if there's more than one row
axes = axes.flatten() if num_charts > 1 else [axes]

# Loop through each optype and create pie charts
for i, (optype, proportions) in enumerate(optype_groups.iterrows()):
    axes[i].pie(proportions, labels=['Non-Cancer', 'Cancer'], autopct='%1.1f%%', startangle=140, colors=['lightblue', 'red'])
    axes[i].set_title(f"{optype}")

# Hide any unused subplots (if there are empty grid spaces)
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

# Adjust layout for readability
plt.tight_layout()
plt.show()
```
